```
import os
import numpy as np


directory = '/content/Script 3. Generating raw data subset/input'
subdirs = {}
num_files = len(os.listdir(directory))
num_subdirs = 4
files_per_dir = num_files/num_subdirs
print(files_per_dir)
print('num dirs = ',num_subdirs)
print('total files = ',num_files)
print('files per dir = ',np.round(files_per_dir,1))
f = 0
for filename in os.listdir(directory):
  target_dir = int(f / files_per_dir)  
  if target_dir in subdirs.keys():
    subdirs[target_dir].append(filename)
  else:
    subdirs[target_dir] = [filename]
  f+=1

s = 0
for key in subdirs.keys():
  n_val = len(subdirs[key])
  print(key, ' : ', n_val)
  s+=n_val

assert(np.round(s,2) == num_files)
  
```

```
%cd '/content/Script 3. Generating raw data subset/input'
count = 1
for key in subdirs.keys():
  folder = '/content/Script_3_Subdir_' + str(key)
  %mkdir '$folder'
  for val in subdirs[key]:
    print('moving ',val)
    print('\tProgress ',100*count/num_files)
    %mv '$val' '$folder'
    count+=1
```
